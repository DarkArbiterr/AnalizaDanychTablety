```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = tablety.drop(columns=['Cena'])
y = tablety['Cena']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=300083)
model.fit(X_train,y_train)
```

```python

print('Wyraz wolny=',model.intercept_)
```

```python
pd.DataFrame(model.coef_, index = model.feature_names_in_, columns = ['wspolczynnik modelu'])
```

```python
print('Cena = ', end='')
for b,n in zip(model.coef_, model.feature_names_in_):
  print(round(b,2), '*', n, '+ ',end='')
  print(round(model.intercept_,2))
```

```python
model.score(X_train,y_train)
```

```python
def adjusted_R2(model, X, y):
  r2 = model.score(X,y)
  n, p = X.shape
  return 1 - (1-r2)*n/(n-p)
adjusted_R2(model, X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from math import sqrt
def ocen_model_regresji(y_true, y_pred, digits = 3):
  print('Średni błąd bezwzględny:', round(mean_absolute_error(y_true,y_pred), digits))
  print('Błąd średniokwadratowy:', round(mean_squared_error(y_true, y_pred),digits))
  print('Pierwiastek błędu średniokwadratowego:',round(sqrt(mean_squared_error(y_true, y_pred)), digits))
  print('Średni bezwzględny błąd procentowy:',round(100*mean_absolute_percentage_error(y_true, y_pred), digits),'%')
```

```python
y_pred = model.predict(X_train)
ocen_model_regresji(y_train, y_pred)
```

```python
y_pred = model.predict(X_test)
ocen_model_regresji(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
def ocen_model_regresji_cv(model, X, y, cv = 5, digits = 3):
  print('Średni błąd bezwzględny: ',round(abs(cross_val_score(model,X,y, scoring = 'neg_mean_absolute_error', cv = cv)).mean(), digits))
  print('Błąd średniokwadratowy: ',round(abs(cross_val_score(model,X,y, scoring = 'neg_mean_squared_error', cv = cv)).mean(), digits))
  print('Pierwiastek błędu średniokwadratowego: ',round(abs(cross_val_score(model,X,y, scoring = 'neg_root_mean_squared_error', cv = cv)).mean(), digits))
  print('Średni bezwzględny błąd procentowy: ',round(100*abs(cross_val_score(model,X,y, scoring = 'neg_mean_absolute_percentage_error', cv = cv)).mean(), digits))
ocen_model_regresji_cv(model, X_train, y_train, cv = len(X_train))
```
